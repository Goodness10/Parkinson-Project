# Feature Selection Techniques and Modelling

USING INFORMATION GAIN

```python
from sklearn.feature_selection import mutual_info_classif

importances = mutual_info_classif(X, y)
feat_importances = pd.Series(importances, df.columns[0:len(df.columns)-1])
top_10_mutual = feat_importances.sort_values(ascending = False).nlargest(10)
top_10_mutual.plot(kind = 'barh', color = 'teal')
plt.xlabel('Correlation')
plt.ylabel('Features')
plt.title('INFORMATION GAIN')
plt.show()
```

```python
#Duplicated X because I want to drop some columns and keep the original X for other purposes
X_dup = X.copy()
```

```python
#picking the top 10 features
X = X[['PPE','spread1', 'MDVP:Jitter(Abs)', 'MDVP:Fo(Hz)', 'spread2', 'MDVP:APQ' , 'MDVP:Flo(Hz)', 'MDVP:Fhi(Hz)', 'HNR' ,'Shimmer:APQ5']]
```

# Modelling

```python
#XGBOOST
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=1234)
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
import sklearn.metrics as metrics

y_pred_xgb=xgb.predict(x_test)
print('The accurcy score for the xgboost models is: ',accuracy_score(y_test, y_pred_xgb) * 100)
```

```python
#confusiion matrix
from sklearn.metrics import confusion_matrix
matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
matrix_xgb
```

```python
#define metrics
y_pred_proba_xgb = xgb.predict_proba(x_test) [::,1]

#Calculate the area under curve to see the model performance
auc_xgb = metrics.roc_auc_score(y_test, y_pred_proba_xgb)
print('The roc_auc_score for xgboost is' ,auc_xgb)
```

```python
false_positive_rate, true_positive_rate, _ = metrics.roc_curve(y_test, y_pred_proba_xgb)

#Create ROC curve
plt.plot(false_positive_rate, true_positive_rate,label="AUC =" + ' ' + str(auc_xgb))
plt.title('ROC Curve')
plt.ylabel('True Positive Rate')
plt.xlabel('false Positive Rate')
plt.legend(loc=4)
```

```python
#Visualize the confusion matrix
sns.heatmap(matrix_xgb, annot=True)
plt.show()
```

```python
from sklearn.metrics import classification_report

Report_xgb = classification_report(y_test, y_pred_xgb)
print(Report_xgb)
```

CATBOOST

```python
#CATBOOST
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier

x_train,x_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=1234)
cat = CatBoostClassifier()
cat.fit(x_train,y_train)
```

```python
y_pred_cat = cat.predict(x_test)
print('The accuracy score for the catboost model is: ',accuracy_score(y_test, y_pred_cat) * 100)
```

```python
#confusiion matrix
from sklearn.metrics import confusion_matrix
matrix_cat = confusion_matrix(y_test, y_pred_cat)
```

```python
#Visualize the confusion matrix
sns.heatmap(matrix_cat, annot=True)
plt.show()
```

```python
#define metrics
y_pred_proba_cat = cat.predict_proba(x_test) [::,1]

#Calculate the area under curve to see the model performance
auc_cat = metrics.roc_auc_score(y_test, y_pred_proba_cat)
print('The roc_auc_score for catboost is' ,auc_cat)
```

```python
false_positive_rate, true_positive_rate, _ = metrics.roc_curve(y_test, y_pred_proba_cat)

#Create ROC curve
plt.plot(false_positive_rate, true_positive_rate,label="AUC =" + ' ' + str(auc_cat))
plt.title('ROC Curve')
plt.ylabel('True Positive Rate')
plt.xlabel('false Positive Rate')
plt.legend(loc=4)
```

```python
from sklearn.metrics import classification_report

Report_cat = classification_report(y_test, y_pred_cat)
print(Report_cat)
```
